```python
import importlib
from pathlib import Path

importlib.reload(importlib.import_module("gen_ceb_query"))

from gen_ceb_query import gen_query

CEB_DIR = Path("/mnt/labstore/bespoke_olap/datasets/ceb/imdb")

template_dict = {}
sql_dict = {}
parameters_dict = {}
for query_name in [
    "1a",
    "2a",
    "2b",
    "2c",
    "3a",
    "3b",
    "4a",
    "5a",
    "6a",
    "7a",
    "8a",
    "9a",
    "9b",
    "10a",
    "11a",
    "11b",
]:
    template, sql, parameters = gen_query(
        CEB_DIR, query_name=query_name, num_queries=10
    )
    template_dict[query_name] = template
    sql_dict[query_name] = sql
    parameters_dict[query_name] = parameters

# write to file
import json

with open("ceb_queries.json", "w") as f:
    json.dump(template_dict, f, indent=4)

# also generate a python file with the templates as a dictionary
with open("ceb_queries.py", "w") as f:
    f.write("ceb_templates = {\n")
    for key, value in template_dict.items():
        safe_value = value.replace('"""', r"\"\"\"")
        f.write(f'    "Q{key}": """{safe_value}""",\n')
    f.write("}\n")
```
