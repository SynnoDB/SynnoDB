```python
path = "/mnt/labstore/bespoke_olap/logs/20260421_115214_xb5idiag_tpch_optim_q1-22_sonnet-4-6_bstorage_20260421_115209_9171.duckdb"

# create a tmp file and copy the duckdb file to it, then connect to it in read-only mode
import tempfile

import duckdb

with tempfile.NamedTemporaryFile(suffix=".duckdb", delete=False) as tmp:
    import shutil

    shutil.copy(path, tmp.name)
    db = duckdb.connect(tmp.name, read_only=True)
```

```python
# inspect the tables in the database
tables = db.execute("SHOW TABLES").fetchall()
print("Tables in the database:")
for table in tables:
    print(table[0])
```

```python
# list columns for table run_metrics
columns = db.execute("PRAGMA table_info('run_metrics')").fetchall()
print("\nColumns in run_metrics:")

apply_patch_columns = [col for col in columns if col[1].startswith("apply_patch")]

for column in sorted(columns, key=lambda x: x[1]):  # sort by column id
    print(column[1])  # column name is in the second position
```

```python
apply_patch_cols_str = ", ".join(f'"{col[1]}"' for col in apply_patch_columns)
```

```python
db.execute(
    f"select step, type, {apply_patch_cols_str} from run_metrics where step = 16"
).df()
```
